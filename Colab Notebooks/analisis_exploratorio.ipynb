```python
#criando nosso Data frame 
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#Visualizando as 5 primeiras linhas 
df.head()
```

```python
#quantidade de linhas e colunas 
df.shape
```

```python
#Verificando os tipos de dado
df.dtypes
```

```python
#Qual o Receita total? 
df["Valor Venda"].sum()
```

```python
#qUAL O CUSTO TOTAL 
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head()
```

```python
round(df["custo"].sum(), 2)
```

```python
#agora que temos a receita e custo e o total, podemos achar o lucro total
#vamos criar uma coluna de lucro que sera Receita -Custo
df["lucro"] = df["Valor Venda"] - df["custo"]
```

```python
df.head(1)
```

```python
#total lucro
round(df["lucro"].sum(),2)
```

```python
#criando uma coluna com total de dias para enviar o produto
df["Tempo_Envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

```python
agora, queremos saber a media do tempo de envio para cada Marca, e para isso precisamos transformar a coluna Tempo_envio em numerica 
```

```python
#Extraendo apenas os dias
df["Tempo_envio"]= (df["Data Envio"]-df["Data Venda"]).dt.days
```

```python
df.head()
```

```python
#Verificando o tipo da coluna Tempo_envio
df["Tempo_envio"].dtype
```

```python
#media do tempo de envio por Marca
df.groupby("Marca")["Tempo_envio"].mean()
```
